#### **Model Training** 

```python
#Normalized YOLO format
from ultralytics import YOLO

model = YOLO("weights/yolov8n.pt")

# Train the model
train_results = model.train(
    data="../sahi/xview-yolo-dataset/YOLO_cfg/xview_yolo.yaml",  # path to dataset YAML
    epochs=150,
    batch = 16,
    imgsz=640,
    multi_scale=True,
    device=0,  # device to run on, i.e. device=0 or device=0,1,2,3 or device=cpu
    #freeze=10,
    lr0 = 0.001,
    lrf = 0.0001,
    dropout = 0.1,
    iou = 0.5,
    close_mosaic = 20,
    cos_lr = True,
    half = True,
    patience=100
)
```
